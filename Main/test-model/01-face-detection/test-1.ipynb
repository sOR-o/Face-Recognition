```
def detect_objects(input_image):

    detected_objects = []
    students = []

    # Fectching the names of detected object
    result: Results = model.predict(input_image)[0]
    for box in result.boxes.boxes:
        detected_objects.append(result.names.get(box[-1].item()))

    # Create a list of detected objects with their indicesc
    for names in enumerate(detected_objects, start = 1):
      students.append(names)

    return str(students)


# Create a Gradio interface for object detection
iface = gr.Interface(
    fn=detect_objects,
    inputs="image",
    outputs="text",
    title="YOLO face detection",
    description="Upload an image to detect objects using YOLO.",
)

iface.launch()
```
